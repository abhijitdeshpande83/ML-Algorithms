# Problem 1 a

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
mu1=[-3,0]
mu2=[3,0]
mu3=[0,3]
cov=[[1,0.75],[0.75,1]]
```

```python
class myKMeans:
    
    def __init__(self, X, k=4):
        self.X = X
        self.K = k
        
    def fit(self):
        self.centroids = X[np.random.choice(len(X), self.K, replace=False)]
        self.intial_centroids = self.centroids
        self.prev_label,  self.labels = None, np.zeros(len(self.X))
        while not np.all(self.labels == self.prev_label) :
            self.prev_label = self.labels
            self.labels = self.predict(self.X)
            self.update_centroid(self.X)
        return self
        
    def predict(self, X):
        return np.apply_along_axis(self.compute_label, 1, X)

    def compute_label(self, x):
        return np.argmin(np.sqrt(np.sum((self.centroids - x)**2, axis=1)))

    def update_centroid(self, X):
        self.centroids = np.array([np.mean(X[self.labels == k], axis=0)  for k in range(self.K)])
```

# Problem 1 b

```python
def plot_cluster(cluster, X, y):
  plt.figure(figsize=(10,8))
  plt.title("Initial centers in red, final centers in green")
  plt.scatter(X[:,0],X[:,1],marker='^',c=y)
  plt.scatter(cluster.centroids[:,0],cluster.centroids[:,1],c='g')
  plt.scatter(cluster.intial_centroids[:,0],cluster.intial_centroids[:,1],c='r')
  plt.show()
```

```python
cluster_1=np.random.multivariate_normal(mu1,cov,300)
cluster_2=np.random.multivariate_normal(mu2,cov,300)
cluster_3=np.random.multivariate_normal(mu3,cov,300)
```

```python
# Plotting actual data
fig=plt.figure(figsize=(8,6))
X=[]
for i in cluster_1:
 X += [i]
for i in cluster_2:
 X += [i]
for i in cluster_3:
 X += [i]
X = np.array(X)
y = [0]*len(cluster_1) + [1]*len(cluster_2) + [2]*len(cluster_3)
plt.scatter(X[:,0], X[:,1],marker='^',c=y)
plt.title("Actual Data")
plt.xlabel("Feature_1")
plt.ylabel("Feature_2")
plt.show()
```

##### Actual vs Predicted

## 2 Clusters

```python
y_actual=[0]*len(cluster_1)+[1]*len(cluster_2)+[2]*len(cluster_3)
cluster = myKMeans(X,2).fit()
plot_cluster(cluster, X, y)
y=cluster.predict(X)
plot_cluster(cluster, X, y)
correct=0
for i in range(len(y)):
  if y_actual[i] == y[i]:
    correct +=1
print("Accuracy for k=2 is:", correct/len(y_actual))
```
